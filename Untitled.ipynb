```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# read img and convert color
img = cv2.imread('Tesla.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# build a transformation matrix
matrix = [[0.5, 0, 0], #x
[0, 1, 0]] #y
t = np.float32(matrix)
# get the sizes
h, w = img.shape[:2]
# transform
img = cv2.warpAffine(img, t, (w, h))
# plot
fig, ax = plt.subplots(1, figsize=(12,8))
ax.axis('off')
plt.imshow(img)
```

```python
def main():

 # Write code here 
    virus = input()
    num_of_samples = int(input())
    samples = []
    blood_len = 0
    virus_len = len(virus)
    i = 0
    j = 0
    
    for _ in range(num_of_samples):
        samples.append(input())
    
    for sample in samples:
        blood_len = len(sample)
        while i < blood_len and j < virus_len:
            if sample[i] == virus[j]:
                i = i + 1
            j = j + 1
        if blood_len == i:
            print('POSITIVE')
        else:
            print('NEGATIVE')
    
main()
```

```python
def main():
    
    def is_prime(num):
        if num > 1:
            for i in range(2,num):
                if (num % i) == 0:
                    break
            else:
                return True
        else:
            return False

    # variables list
    num_ranges = []

    # take them inputs
    inputs = int(input())
    for _ in range(inputs):
        num_ranges.append(input())

    # convert each input to a tuple of integers
    num_ranges = [(int(num.split()[0]), int(num.split()[1])) for num in num_ranges]

    # each tuple list of numbers between given range
    numbers = [[i for i in range(L, R+1)] for L, R in num_ranges]
    
    # delete endpoints in the number list if they are not prime numbers
    for num_list in numbers:
        while (len(num_list) != 0) and (not is_prime(num_list[0]) or not is_prime(num_list[-1])):
            if len(num_list) and not is_prime(num_list[0]):
                del num_list[0]
            if len(num_list) != 0 and not is_prime(num_list[-1]):
                del num_list[-1]
            if len(num_list) != 0 and num_list[0] == 1:
                del num_list[0]
    
    # code in the business rules
    for num_list in numbers:
        if len(num_list) == 0:
            print(-1)
        elif len(num_list) == 1:
            print(0)
        else:
            print(num_list[-1] - num_list[0])
    
main()
```
