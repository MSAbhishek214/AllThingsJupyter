```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
dataset = pd.read_csv('telco1.csv')
dataset.head()
```

```python
dataset.info()
```

```python
nan_value = float('NaN')
dataset.replace(' ', nan_value, inplace=True)
dataset.dropna(inplace=True)
```

```python
dataset.info()
```

```python
print(dataset.dtypes)
```

```python
print(dataset.dtypes)
```

```python
sns.pairplot(dataset)
```

```python
sns.distplot(dataset['tenure'])
```

```python
sns.heatmap(dataset.corr(), annot=True, linewidths=2)
```

```python
dataset.columns
```

```python
X=dataset[['sales', 'resale', 'engine_s', 'horsepow', 'wheelbas', 'width',
       'length', 'curb_wgt', 'fuel_cap', 'mpg']]
```

```python
y=dataset['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
print(lr.intercept_)
```

```python
lr.coef_
```

```python
cdf = pd.DataFrame(lr.coef_, X_train.columns, columns=['Coeff'])
```

```python
cdf
```

```python
predictions = lr.predict(X_test)
```

```python
plt.scatter(y_test, predictions)
```

```python
sns.distplot((y_test - predictions))
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test, predictions)
```

```python
MSE = metrics.mean_squared_error(y_test, predictions)
MSE
```

```python
np.sqrt(MSE)
```

The given data set is not fully normalized hence there is some noise in the curve shown which is not looking exactly like a bell curve.
